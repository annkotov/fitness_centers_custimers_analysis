## Шаг 3. Модель прогнозирования оттока пользователей

Построим модель прогнозирования оттока пользователей бинарной классификации пользователей, где целевой признак — факт оттока пользователя в следующем месяце:
   - Разобьем данные на обучающую и валидационную выборку функцией train_test_split().
   - Обучим модель на train-выборке двумя способами:
        - логистической регрессией
        - случайным лесом
   - Оценим метрики accuracy, precision и recall для обеих моделей на валидационной выборке и по ним определим, какая модель показала себя лучше на основании метрик


```python
# Выделим целевую переменную и признаки
X = df.drop('churn', axis = 1)
y = df['churn']
```

```python
# Разделим данные на обучающую 80% 'train' и валидационную 20% 'test' выборку
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Стандартизируем данные
scaler = StandardScaler()
scaler.fit(X_train)
X_train_st = scaler.transform(X_train)
X_test_st =  scaler.transform(X_test)
```

```python
# обозначим функцию печати метрик
def print_all_metrics(y_true, y_pred, y_proba, title='Метрики классификации'):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))

```

### Модель логистической регрессии

```python
# Зададим алгоритм для модели логистической регрессии.
lr_model = LogisticRegression(random_state=0)

# Обучим модель.
lr_model.fit(X_train_st, y_train)

# Воспользуемся уже обученной моделью, чтобы сделать прогнозы.
lr_predictions = lr_model.predict(X_test_st)
lr_probabilities = lr_model.predict_proba(X_test_st)[:,1]

print_all_metrics(
    y_test,
    lr_predictions,
    lr_probabilities,
    title='\033[1mМетрики для модели логистической регрессии:\033[0m',
)
```

### Модель случайного леса

```python
rf_model = RandomForestClassifier(n_estimators = 100,random_state = 0) # Ваш код здесь
# обучим модель случайного леса
rf_model.fit(X_train_st, y_train)
# воспользуемся уже обученной моделью, чтобы сделать прогнозы
rf_predictions = rf_model.predict(X_test_st)# Ваш код здесь
rf_probabilities = rf_model.predict_proba(X_test_st)[:, 1]# Ваш код здесь
# выведем все метрики
print_all_metrics(
    y_test,
    rf_predictions,
    rf_probabilities,
    title='\033[1mМетрики для модели случайного леса:\033[0m'
)

```

## Заключение

Отток клиентов - распространенная проблема многих сервисов предоставляющих услуги разного рода, в том числе, конечно же, фитнес клубов. Уход клиентов опасен со всех сторон: очевидно, что привлечение и удержание новых клиентов затратнее с точки зрения финансовых вложений в маркетинг, но есть еще и более глубокая проблема. Постоянная ротация клиентской базы (приход и уход клиентов) клуба оказывает непосредственного вляние на атмосферу клуба и как следствие - препятствует удержанию уже привлеченных клиентов, что может замкнуть круг и поставить под удар функционирование предприятия как такового.

Условно посетителей фитнес-клубов и других спортивных организаций можно разделить на две большие группы:
   - постоянные: те, кто соблюдают режим вне зависимотсти от мероприятий, времени года, праздников и т.д. Для этих клиентов факт оттока скорее будет носить сезонный характер: они могут изменить вид активности на аутдор-тренировки, например, летом, и вернуться вместе с началом непогоды, либо уехать в длительный отпуск и после него вернуться в привычный тренировочный режим. Окончательный отток из клуба для них скорее будет связан со сменой места работы и/или жительства, вследствие чего посещение данного фитнес клуба будет не удобно с точки зрения временных затрат, либо получение травмы и как следствие, необходимость индивидуально составленной программы специальных тренировок, либо смена направления физической активности;
   - сезонные (они же "подснежники"): это большая группа клиентов, которые начинают активный тренинг с приходом весеннего сезона с целью "привести себя в форму к лету/к праздникам/к отпуску". Как правило, они настроены на быстрый результат и менее подкованы в вопросе организации тренировочного процесса
   
Стратегии по удержанию этих групп клиентов должны бать разными: для второй группы необходимо предварительно работать в сторону перехода их в первую, группу постоянных мотивированных клиентов. Таким образом, стратегию работы по борьбе с оттоком клиентов можно разделить на три этапа:

1. Привлечение новых клиентов
2. Перевод клиентов из группы новичков - сезонников в группу регулярно занимающихся (Для части клиентов этот шаг может быть пропущен)
3. Удержание постоянных клиентов

На основании всего вышеизложенного, рекомендации по стратегиям работы с данными группами клиентов будут следующие.


### Привлечение новых клиентов

1. Ориентация на потенциальных клиентов, проживающих или работающих в непосредственной близости с клубом:
    - Локальная реклама (Геотаргетинг)
    - При открытии новых филлиалов ориентироваться на близость бизнес-центров, предприятий с учетом среднего возраста сотрудников и средней заработной платы (ориентация на молодежь)
2. Привлечение клиентов разных возрастных групп: введение новых направлений тренировок
    - Открытие детских групповых занятий (ориентация на молодых родителей: возможность тренироваться параллельно с ребенком)
    - Открытие новых групп для людей старшего возраста (с учетом возрастных особенностей)
3. Введение новых типов абонементов (как следствие пункта 2):
    - Семейный абонемент
    - Дневной абонемент со скидкой (позволит привлечь пенсионеров, активность которых последние годы набирает обороты плюс обеспечить посещаемость клуба в нетипичное время)
    - Детский абонемент (Как вариант - родитель приводит ребенка на занятие, а сам идет в салон красоты на массаж или в бар при клубе. В таком случае продумать систему лояльности)
4. Повышение квалификации сотрудников и/или расширение штата для обеспечения открытия новых направлений
5. Поддержка и расширение партнерских программ и акции "приведи друга" как успешно работающих методов привлечения

### Конвертация клиентов в группу регулярно занимающихся

1. Личное взаимодействие с каждым клиентом
   - Введение новоприбывших в клуб (интеграция): выгодные предложения на курс вводных тренировок для новичков, активное участие в их спортивной жизни
   - Ведение статистики их посещений и фиксация прогресса (например, ьрекер тренировок в приложении клуба) - с одной стороны, для контроля со стороны тренера и своевременных напоминаниях и корректировках тренировочного процесса, с другой - для наглядной демонстрации прогресса самому клиенту
2. Обеспечение дружественной комфортной атмосферы в клубе
   - Проведение праздников, тематических мероприятий в клубе
   - Проведение внутриклубных соревнований и поощрение победителей
   - Вовлеченность в жизнь клиентов: поздравления с праздниками, отслеживание статистики их посещений и ненавязчивые напоминания в случае их длительного отсутствия (например, звонок или смс-оповещение в случае долгого отсутствия с предложением заморозки абонемента)

### Удержание постоянных клиентов

Рекомендации по удержанию клиентов в целом схожи со стратегриями по привлечению и конвертации клиентов и будет заключаться в первую очередь с отлавливанием "первых звоночков, предваряющих уход" (то есть, направить работу на предотвращение оттока как массового явления):

1. Анализ статистики средненедельной посещаемости. В случае стабильного снижения этого показателя - ненавязчиво выяснить приины и возможно:
    - предложить гибкие условия заморозки абонемента
    - предложить разнообразить тренировочный процесс (например, рассказать о новых направлениях)
    - мотивировать возобновлять режим тернировок: предложить скидку на комплекс услуг от персонального тренера, устроить розыгрыш/соревнования (для подготовки к ним придется снова войти в режим), разработать новую систему лояльности в зависимости от стажа тренировок в клубе
2. Так как постоянные клиенты - часто уже люди искушенные в вопросах тренинга, проводить голосования в целях выявления потребностей в новых направлениях. Например, открытие новых популярных или набирающих популярность направлений позволят увеличить посещаемость как групповых занятий (например, организовать группы по направлению боевых искуств, скалолазания, акробатики, танцам), так и самостоятельных посещений (с целью ОФП или СФП)
3. Как следствие п.2 - постоянное повышение квалификации тренерского состава и разумное расширение направлений
